```python
print("IRIS - regression")
print("Предсказываем Sepal length по остальным признакам")
```

```python
X, y = iris_loader()

y = X[:, 0].astype(float)
X = np.delete(X, 0, axis=1) 

test_size=0.2
random_state=None

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)    

regression = NeuralNetwork(activation_func=sigmoid, activation_derivative=sigmoid_der, optimizer=momentum_sgd, loss_function='mse')

input_size = X_train.shape[1]
output_size = 1  

regression.add_layer(input_size, 64)
regression.add_layer(64, 512)
regression.add_layer(512, 1024)
regression.add_layer(1024, 512)
regression.add_layer(512, 128)
regression.add_layer(128, 64)
regression.add_layer(64, output_size, activation_func = linear, activation_derivative = linear_der)

regression.train(X_train, y_train, epochs=200, learning_rate=0.0001, batch_size=32)

y_pred = regression.forward(X_test) 

loss = mse_loss(y_test, y_pred)
print(f"Test Loss: {loss:.4f}")

print(f"Predicted values: {y_pred.ravel()}")
print(f"True values: {y_test.ravel()}")
```

```python
print("IRIS - classification")
```

```python
X, y = iris_loader()
y = one_hot_encode(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
classifier = NeuralNetwork(activation_func=relu, activation_derivative=relu_der, optimizer=sign_sgd, loss_function='cross_entropy')

input_size = X_train.shape[1]
output_size = y_train.shape[1]

classifier.add_layer(input_size, 8)  
classifier.add_layer(8, output_size)

classifier.train(X_train, y_train, epochs=300, learning_rate=0.0001, batch_size=16)
y_pred = classifier.forward(X_test)

y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

accuracy = np.mean(y_pred_classes == y_test_classes)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
X,y = mnist_loader()
y = one_hot_encode(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
classifier = NeuralNetwork(activation_func=relu, activation_derivative=relu_der, optimizer=sgd, loss_function='cross_entropy')
input_size = X_train.shape[1]
output_size = y_train.shape[1]
classifier.add_layer(input_size, 128)
classifier.add_layer(128, output_size)

classifier.train(X_train, y_train, epochs=200, learning_rate=0.001, batch_size=32)
y_pred = classifier.forward(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

accuracy = np.mean(y_pred_classes == y_test_classes)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
